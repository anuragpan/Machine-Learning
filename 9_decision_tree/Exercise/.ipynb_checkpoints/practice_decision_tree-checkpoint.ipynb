```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
y = df.Survived
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
obj1 = LabelEncoder()
obj2 = LabelEncoder()
obj3 = LabelEncoder()
obj4 = LabelEncoder()
```

```python
df['Pclass_n'] = obj1.fit_transform(df.Pclass)
df['Sex_n'] = obj1.fit_transform(df.Sex)
df['Age_n'] = obj1.fit_transform(df.Age)
df['Fare_n'] = obj1.fit_transform(df.Fare)
df.head()
```

```python
df_n = df.drop(['Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked','PassengerId','Survived','Pclass'],axis=1)
df_n.head()
```

```python
df_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(df_n,y)
```

```python
out[] above tell us about what were the parameters on which our model was trained we can see by: 
```

```python
model.criterion
```

```python
from sklearn.externals import joblib
joblib.dump(model,'model_joblib')
```

```python
model_joblib = joblib.load('model_joblib')
```

```python
model_joblib.score(df_n,y)
```

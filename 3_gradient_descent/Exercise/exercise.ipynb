```python
%matplotlib inline
def gd(x,y):
    m_curr = b_curr = 0
    n = len(x)
    rate = 0.01
    iter = 100
    for i in range(iter):
        y_pred = m_curr*x + b_curr
        cost = (1/n)*sum([value**2 for value in (y-y_pred)])
        print(m_curr,b_curr,i,cost)
        md = -1*(2/n)*sum(x*(y - y_pred ))
        bd = -1*(2/n)*sum(y-y_pred)
        m_curr = m_curr - rate*md
        
        b_curr = b_curr - rate*bd
        
        
```
